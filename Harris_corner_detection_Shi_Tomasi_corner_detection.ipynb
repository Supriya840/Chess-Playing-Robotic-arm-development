```python
real_board = cv2.imread('im_7.jpg')
real_board = cv2.cvtColor(real_board, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_board)
```

```python
gray_real_board = cv2.cvtColor(real_board, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_real_board, cmap='gray')
```

```python
gray_new_board
```

```python
gray = np.float32(gray_new_board)
```

Harris corner detection method

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.03)
```

```python
dst = cv2.dilate(dst, None)
```

```python
new_board[dst > 0.01 * dst.max()] = [255, 0, 0]
```

```python
plt.imshow(new_board)
```

```python
gray = np.float32(gray_real_board)

dst = cv2.cornerHarris(src= gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
dst
```

```python
real_board[dst > 0.01 * dst.max()] = [255, 0, 0]
```

```python
plt.imshow(real_board)
```

Shi Tomasi corner detection method


```python
corners = cv2.goodFeaturesToTrack(gray_new_board, 5, 0.01, 10)
```

```python
corners
```

```python
corners = np.int32(corners)
```

```python
corners
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(new_board, (x, y), 3, (255, 0, 0), -1)
```

```python
plt.imshow(new_board)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_board, 100, 0.01, 10)
```

```python
corners = np.int32(corners)
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(real_board, (x, y), 3, (255, 0, 0), -1)
```

```python
plt.imshow(real_board)
```
